```python
import cv2
import matplotlib.pyplot as plt # 用来绘图 
import numpy as np # 矩阵和向量运算
```

```python
# 读取图片
img = cv2.imread("car.jpg")
```

```python
# 封装用于显示图片
def imgshow(name):
    plt.imshow(name)
    plt.show()
imgshow(img)
```

```python
# 也可以使用opencv自带的打开图像的函数 不过并不好用
# 因为它会从新打开一个窗口
cv2.imshow("窗口的名字",img)
cv2.waitKey(10000) 
cv2.destroyAllWindows()
```
